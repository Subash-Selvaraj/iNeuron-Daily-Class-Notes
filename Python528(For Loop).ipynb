##### Q6 try to extract all numeric data it may be a part if dict key and value

```python
l1 = []
for i in range(len(l)):
    if type(l[i])==list or type(l[i]) ==tuple or type(l[i]) ==set:
        for j in l[i]:
            if type(j)==int:
                l1.append(j)
    if type(l[i])==dict:
        for k,m in l[i].items():
            if type(k)==int or type(m)==int:
                l1.append(k)
                l1.append(m)
print(l1)
```
