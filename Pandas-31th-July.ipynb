### Merge

```python
data4 = {
    "emp_id":[101,102,103,104],
    "salary":[234,345,300,400],
    "pf":[23,34,56,78]
}
```

```python
data5 ={
    "emp_id":[101,102,104,105],
    "mob_no":[3434,5656,767,878],
    'house_no':[345,345,345,56]
}
```

```python
df4 = pd.DataFrame(data4)
```

```python
df4
```

```python
df5 = pd.DataFrame(data5)
```

```python
df5
```

```python
#what is the house number and salary of emp_id 101 here we have two tables those are having hous no and salary details so here we have to merge both tables
#merge functions will first look the common values of both tables and it trying to merege the datas(inner merge)
pd.merge(df4,df5)
```

```python
#left merge (same as left join in sql) #get the all the data from left table 
#--and what ever data we have common in left and right it will retuen the same

pd.merge(df4,df5,how = 'left')
```

```python
#right merge (same as left join in sql) #get the all the data from right table 
#--and what ever data we have common in right and left it will retuen the same

pd.merge(df4,df5,how = 'right')
```

```python
pd.merge(df4,df5,how = 'right',on='emp_id') #we can define the column for comparing the two tables
```

```python
##load dataset
dfs = pd.read_csv(".\DataSets\data fsds\sales_data_final.csv")
dfs.head()
```

```python
def profit_flag(a):
    if a>=0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(0)
```

```python
#create column to show the profit flag is negative or positive based on the profit columns values
dfs['profit_flag']=dfs['profit'].apply(profit_flag)
```

```python
dfs.head(10)
```

```python
#create a colum tpo show the length of customer name
dfs['len_cus_name'] =dfs['customer_name'].apply(len)
```

```python
dfs
```

```python
#categrice the quantity into three segements(high(20+,medium(11-20),low(<10)
def quantity_flag(a):
    if a <10:
        return 'Low'
    elif a>=10 and a<2:
        return 'Medium'
    else:
        return 'Low'
```

```python
quantity_flag(10)
```

```python
#apply the above in to the table
dfs['quantity_flag'] = dfs['quantity'].apply(quantity_flag)
```

```python
dfs
```

```python
#make square of the quantity by using lambda function
dfs['square_quantity']= dfs['quantity'].apply(lambda a :a**2)
```
