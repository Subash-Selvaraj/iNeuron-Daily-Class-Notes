```python
def square(n):
    for i in range(n):
        sq = i*i
        print(sq)
```

```python
square(10)
```

```python
def squ_fun(n):
    sq = [i*i for i in range(n)]
    return sq
```

```python
squ_fun(10)
```

```python
#generator
def squ_fun1(n):
    sq = [i*i for i in range(n)]
    yield sq
```

```python
squ_fun1(10)
```

```python
#function is an object so then we have to iterare tge values by using loops
for i in squ_fun1(10):
    print(i)
```

```python
for i in squ_fun1(100000):
    print(i)
```

```python
for i in squ_fun1(200000):
    print(i)
```

```python
#fib
def genfib(n):
    a = 1
    b = 1
    l = []
    for i in range(n):
        l.append(a)
        a,b = b,a+b
    return l
```

```python
genfib(10)
```

```python
#fib by using genrator
def genfib(n):
    a = 1
    b = 1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
for i in genfib(10):
    print(i,end=' ')
```

# MAP

```python
#map will take 2 arguments
def sq(n):
    return n**2
```

```python
map(sq,l) #list is iterble which is given as l and map function will do iterate each and very element and it will perform based on the function which we have given in first argument
```

```python
list(map(sq,l)) #first argument is function second argument is iterable object like string or list
```

```python
#by using lambda function
list(map(lambda x:x**2,l))
```

```python
#store the every data from list in str format
list(map(str,l))
```

```python
list(map(lambda x:str(x),l))
```

```python
def str_func(n):
    return str(n)
```

```python
list(map(str_func,l))
```

# Filter

```python
l = [1,2,3,4,5]
```

```python
l1 = [i for i in l if i%2==0] #get all even  number from list
```

```python
l1
```

```python
#the above requirement we can use filter function
filter(lambda x:x%2==0,l)
```

```python
list(filter(lambda x:x%2==0,l))
```

```python
def even(n):
    if n%2==0:
        return n
```

```python
list(filter(even,l))
```
